# Relatório  de Análise III - Seleção e exportação

## Imóveis Residenciais

```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel.csv', sep = ';')
```

```python
dados.head(10)
```

```python
list(dados['Tipo'].drop_duplicates())
```

```python
dados.query('Tipo=="Quitinete" or Tipo=="Casa" or Tipo=="Apartamento" or Tipo=="Casa de Condomínio" or Tipo=="Casa de Vila"')
```

```python
residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa de Vila']
```

```python
residencial
```

```python
dados.head(10)
```

```python
sel = dados['Tipo'].isin(residencial)
dados_residencial = dados[sel]
dados_residencial
```
