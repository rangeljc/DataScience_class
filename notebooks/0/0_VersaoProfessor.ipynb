## **Queries nos dados**

```python
dados.head()
```

```python
#Selecione os carros com motor a diesel.
selecao = dados['Motor'] == 'Motor Diesel'
selecao
```

```python
dados[selecao]
```

```python
#Selecione os carros com motor a diesel e com valor menor que 80000 reais
selecao = (dados['Motor'] == 'Motor Diesel') & (dados['Valor'] < 80000) #& - > AND
dados[selecao]
```

```python
#Selecione os carros com motor a diesel ou com valor menor que 80000 reais
selecao = (dados['Motor'] == 'Motor Diesel') | (dados['Valor'] < 80000) #| - > OR
dados[selecao][['Valor', 'Ano']]
```

**Exemplo**: Calcular a quilometragem média anual

```python
dados.head() #2019
```

```python
44410.0/(2019-2003)
```

```python
dados.query('Ano==2019')
```

```python
ano_atual = 2019

for id, row in dados.iterrows():
  anos_rodando = ano_atual - row['Ano'] #denominador==0
  if anos_rodando!=0:
    dados.loc[id, 'Km_media'] = row['Quilometragem'] / anos_rodando
  else:
    dados.loc[id, 'Km_media'] = 0

dados.head()
```
